```python
v1 = np.array([1, 2, 3, 4])
v2 = np.array([0, -4, -3, -1])

# method 1 
dp1 = sum(np.multiply(v1,v2))

# method 2
dp2 = np.dot(v1,v2)

# method 3
dp3 = np.matmul(v1,v2)

# method4
dp4 = 0

for i in range(0, len(v1)):
    dp4 = dp4 + v1[i]*v2[i]
```

```python
# distributive
n = 10
a = np.random.randn(n)
b = np.random.randn(n)
c = np.random.randn(n)


res1 = np.dot(a, (b+c))
res2 = np.dot(a,b) + np.dot(a,c)

print(res1, res2)

```

```python
# associative
n = 5
a = np.random.randn(n)
b = np.random.randn(n)
c = np.random.randn(n)


res1 = np.dot(a, np.dot(b,c))
res2 = np.dot(np.dot(a,b),c)

print(res1, res2)
print("Not the same because dot product is not associative")
```

```python
v1 = np.array([2, 4, -3])
v2 = np.array([0, -3, -3])
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.plot([0, v1[0]], [0, v1[1]], [0, v1[2]], 'b')
ax.plot([0, v2[0]], [0, v2[1]], [0, v2[2]], 'r')

plt.axis((-6, 6, -6, 6))
plt.title("Angle between vectors: %s rad." % ang)
plt.show()
```

```python
# outer product
v1 = np.array([1, 2, 3])
v2 = np.array([-1, 0, 1])
a = np.outer(v1,v2)
b = np.outer(v2,v1)
print(a)
print(b)

#terrible programming but helps conceptually
op = np.zeros((len(v1), len(v1)))
for i in range(0, len(v1)):
    for j in range(0, len(v2)):
        op[i,j]= v1[i] * v2[j]
print(op)
```

```python
# cross product
v1 = [-3, 2 , 5]
v2 = [4, -3, 0]

v3a = np.cross(v1,v2)

v3b = [ [v1[1]*v2[2] -  v1[2]*v2[1]],
        [v1[2]*v2[0] -  v1[0]*v2[2]],
        [v1[0]*v2[1] -  v1[1]*v2[0]] ]
print(v3a, v3b)

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')

xx, yy = np.meshgrid(np.linspace(-10, 10, 10), np.linspace(-10, 10, 10))
z1 = (-v3a[0]**xx - v3a[1]**yy)/v3a[2]
ax.plot_surface(xx,yy,z1)

ax.plot([0, v1[0]], [0, v1[1]], [0, v1[2]], 'k')
ax.plot([0, v2[0]], [0, v2[1]], [0, v2[2]], 'k')
ax.plot([0, v3a[0]], [0, v3a[1]], [0, v3a[2]], 'r')

ax.view_init(azim=150, elev=45)
plt.show()
```
