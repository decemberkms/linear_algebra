```python
import numpy as np
import pandas as pd
```

```python
# square vs rectangular
S = np.random.randn(5,5)
R = np.random.randn(5,2)
print(S), print(" ")
print(R)

# identity
I = np.eye(3)
print(I)

# zeros
Z = np.zeros((4,4))
print(Z)

# diagonal
D = np.diag([1,2,3,5,2])
print(D)

# trianguler
S = np.random.randn(5, 5)
U = np.triu(S)
L = np.tril(S)
print(S)
print(U)
print(L)

#concatenate 
A = np.random.randn(3, 2)
B = np.random.randn(3, 4)
C = np.concatenate((A,B), axis = 1)

print(C)
           
```

```python
A = np.random.randn(5,4)
C = np.random.randn(5,4)
A + C

#shifting
A = np.random.randn(5,5)
print(A)
A = A + 1*np.eye(5)
print(A)
```

```python
M = np.array([[1,2],[2,5]])
s = 2
print(M*s)
print(s*M)
```

```python
M = np.array([[1,2,3],[2,3,4]])
print(M)
print(M.T)
print(M.T.T)
print(np.transpose(M))
```

```python
C = np.array([4 +1j,3, 2-4j])
print(C)
print(C.T)
print(np.transpose(C))
print(np.matrix(C).H) # hermitian works only this way
```
