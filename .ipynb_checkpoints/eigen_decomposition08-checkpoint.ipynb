```python
A = [[1,5], [2,4]]

eigvals = np.linalg.eig(A)

print(eigvals[0]) # eigen value
print(eigvals[1]) # eigen vectors

v1 = np.array([1,1])
v2 = np.random.randn(2,1)
v2 = v2/np.linalg.norm(v2)

Av1 = A@v1
Av2 = A@v2


print(v1)
print(v2)
# plot the vectors and Av
plt.plot([0,v1[0]] ,[0,v1[1]],'r')
plt.plot([0,Av1[0]],[0,Av1[1]],'r--')
plt.plot([0,v2[0][0]] ,[0,v2[1][0]],'k')
plt.plot([0,Av2[0][0]],[0,Av2[1][0]],'k--')

plt.axis([-8,8,-8,8])
plt.show()

# singular matrix => determinant  is  zero
```

```python
# eigen vector

A = [[1, 2], [2, 1]]

evals, evecs = np.linalg.eig(A)
print(evals)
print(evecs)

mag_v1 = np.sqrt( np.sum(np.square(evecs[:,0])))
mag_v2 = np.sqrt( np.sum(np.square(evecs[:,1])))
print(mag_v1, mag_v2)

plt.plot([0, evecs[0,0]], [0, evecs[1,0]], 'r', label = 'v1')
plt.plot([0, evecs[0,1]], [0, evecs[1,1]], 'k', label = 'v2')
plt.axis([-1,1,-1,1])

plt.show()
```

```python

A = np.round(10 * np.random.randn(4, 4))
A = A.T@A

evals, evecs = np.linalg.eig(A)

Ap = evecs @ np.diag(evals) @ np.linalg.inv(evecs)

plt.subplot(121)
plt.imshow(A)
plt.axis("off")
plt.title("A")

plt.subplot(122)
plt.imshow(Ap)

plt.show()

recondiff = A - Ap
recondiff
```

```python
# matrix powers

A = np.random.randn(2, 2)

# compute matrix power directly
print(np.linalg.matrix_power(A,3))
print(A@A@A)
print("")

D, V = np.linalg.eig(A)
D = np.diag(D)
print(V @ np.linalg.matrix_power(D, 3)@ np.linalg.inv(V))
```

```python
# eigenvalues of A and A^3

# create symetric matrix 
A = np.random.randn(3, 3)
A = A @ A.T

D, V = np.linalg.eig(A)
D3, V3 = np.linalg.eig(A@A@A)

print(V, "\n\n", V3)

plt.subplot(221)
plt.imshow(V)
plt.axis("off")
plt.title("evecs of A")

plt.subplot(223)
plt.imshow(V3)
plt.axis("off")
plt.title("evecs of A")

```

```python
## plot the eigenvectors and eigenvalues

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')

# plot eigenvectors of A
ax.plot([0,V[0,0]],[0,V[1,0]],[0,V[2,0]],'r')
ax.plot([0,V[0,1]],[0,V[1,1]],[0,V[2,1]],'k')
ax.plot([0,V[0,2]],[0,V[1,2]],[0,V[2,2]],'b')

# plot eigenvectors of A^3
ax.plot([0,V3[0,0]],[0,V3[1,0]],[0,V3[2,0]],'r--')
ax.plot([0,V3[0,1]],[0,V3[1,1]],[0,V3[2,1]],'k--')
ax.plot([0,V3[0,2]],[0,V3[1,2]],[0,V3[2,2]],'b--')

plt.show()

## show the eigenvalues
plt.plot([1,2,3],D,'bs-',label='A')
plt.plot([1.1,2.1,3.1],D3,'rs-',label='A^3')
plt.title('Eigenvalues')
plt.legend()
plt.show()
```

```python
# (A-B)V = labmda v
# (A^2 - AB - BA + B^2)v = lambda^2 v

A,B = np.random.randn(5,5),np.random.randn(5,5)
A,B = A@A.T, B@B.T # to remove imaginary part

AminusB = A - B
Daminusb,Vaminusb = np.linalg.eig(AminusB)

AminusBsquared = A@A - A@B - B@A + B@B
Daminusbsquared,Vaminusbsquared = np.linalg.eig(AminusBsquared)

print(Vaminusb)
print(Vaminusbsquared)
print("")
print(Daminusb**2, " \n", Daminusbsquared)
```

```python
# create an m x m symmetric matrix 5 x 5 take eigendecomp

m = 5
A = np.random.randn(5, 5)
A = np.round(10 * A.T@A)
print(A)
D, V = np.linalg.eig(A)

# convert to column vector
v = V[:,2]
print(v)
print(np.reshape(V[:,2], (m, 1))) # how to make it a vector
```

```python
op = np.outer(v,v)
print(op), print(" ")

v = np.reshape(V[:, 2], (m,1))
print(v * v.T), print(" ") # like math

# check the norm of outer product
print(np.linalg.norm(op))
```

```python
print(np.linalg.norm(v*D[2]*v.T))
print(D[2])
```

```python
# reconstruct A by summer over outer products

Arecon = np.zeros((m, m))

for i in range(m):
    v = np.reshape(V[:,i], (m,1))
    Arecon += v*D[i]*v.T
    print(Arecon)
    print(np.linalg.matrix_rank(Arecon))

print(A); print("")
print(Arecon); print("")
print(np.round(A - Arecon, 1)); print("")


```
