```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import math
from sympy import *

import numpy as np
import matplotlib.pyplot as plt
from sympy import *
import scipy.io as sio
from mpl_toolkits.mplot3d import Axes3D
```

```python
A = [[1,5], [2,4]]

eigvals = np.linalg.eig(A)

print(eigvals[0]) # eigen value
print(eigvals[1]) # eigen vectors

v1 = np.array([1,1])
v2 = np.random.randn(2,1)
v2 = v2/np.linalg.norm(v2)

Av1 = A@v1
Av2 = A@v2


print(v1)
print(v2)
# plot the vectors and Av
plt.plot([0,v1[0]] ,[0,v1[1]],'r')
plt.plot([0,Av1[0]],[0,Av1[1]],'r--')
plt.plot([0,v2[0][0]] ,[0,v2[1][0]],'k')
plt.plot([0,Av2[0][0]],[0,Av2[1][0]],'k--')

plt.axis([-8,8,-8,8])
plt.show()

# singular matrix => determinant  is  zero
```
