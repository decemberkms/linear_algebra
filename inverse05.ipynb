```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import math
```

```python
# diagonal matrix inverse
A = np.array([[2,0], [0,3]])
A = np.diag(np.arange(1,6))
print(A), print(" ")
print(np.linalg.inv(A))
```

```python
m = 6
n = 3
A = np.random.randn(m, n)
AtA = A.T@A
AAt = A@A.T
print("Shape of A^TA: ", np.shape(AtA))
print("rank of A^TA: ", np.linalg.matrix_rank(AtA))
print("Shape of AA^T: ", np.shape(AAt))
print("rank of AA^T: ", np.linalg.matrix_rank(AAt))

# right inverse
Aleft = np.linalg.inv(AtA)@A.T

# left inverse
Aright = A.T@np.linalg.inv(AAt)

# test
I_left = Aleft @ A
I_right = A @ Aright

print(I_left)
print(I_right)

# test using inverse function
AtA_inv = np.linalg.inv(AtA)
I_AtA = AtA_inv @ AtA

AAt_inv = np.linalg.inv(AAt)
I_AAt = AAt_inv @ AAt

print(I_AtA)
print(I_AAt)
```

```python
# Pseudo inverse
pseudoInV = np.linalg.pinv(A)

plt.subplot(131)
plt.imshow(A),  plt.axis('off')
plt.title('A')


plt.subplot(132)
plt.imshow(pseudoInV),  plt.axis('off')
plt.title('pseudoInV of A')


plt.subplot(133)
plt.imshow(pseudoInV @ A),  plt.axis('off')
plt.title('A$^*$A')
plt.show()


print(pseudoInV @ A)
```

```python
# 50 50 matrix
n = 50
A = np.random.randn(n, n)
A[:, n -1] = A[:, n-2]
print('rank = ', np.linalg.matrix_rank(A))

Ai = np.linalg.pinv(A)
```

```python
plt.subplot(221)
plt.imshow(A),  plt.axis('off')
plt.title('A')


plt.subplot(222)
plt.imshow(Ai),  plt.axis('off')
plt.title('A$^*$')


plt.subplot(223)
plt.imshow(Ai @ A),  plt.axis('off')
plt.title('A$^*$A')
plt.show()

plt.subplot(224)
plt.imshow(A @ Ai),  plt.axis('off')
plt.title('AA$^*$')
plt.show()
```

```python
# pseudo inverse is the same as the true inverse if matrix is invertible?
m = 5

A = np.random.randn(5,5)

Ainv = np.linalg.inv(A)
Apinv = np.linalg.pinv(A)
print("Are they the same? \n", np.round(Ainv - Apinv, 5))
print("A @ Ainv\n", A @ Ainv)
print("A @ Apinv\n", A @ Apinv)
```
